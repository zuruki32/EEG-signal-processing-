```python
#MAKING DATALIST FILE FROM DATA WITH PREPROCCEING(median filtering AND )
import pandas as pd
import numpy as np
import mne
import os
from pygsp import graphs, utils
from scipy.spatial import distance_matrix
from scipy.stats import entropy
import networkx as nx
from sklearn.cluster import spectral_clustering
from sklearn.preprocessing import StandardScaler

dir_path = r'/content/drive/MyDrive/dataseta'
label_file_path = '/content/drive/MyDrive/alzimer_result/participants.csv'
output_dir = "/content/drive/MyDrive/dataseta_processed"
file_list = [entry.path for entry in os.scandir(dir_path) if entry.is_file() and entry.path.endswith(".set")]

n_files = len(file_list)
if n_files == 0:
    raise ValueError(f"No .set files found in directory {dir_path}.")

print(f'Found {n_files} .set files.')

# Read labels from CSV file
labels_df = pd.read_csv(label_file_path)

# Define mapping for classes
class_mapping = {'A': 0, 'C': 1, 'F': 2}

# Initialize a list to store processed data
data_list = []

# Iterate over the EEG data files
for i, file in enumerate(file_list):
    # Extract participant ID from the file name
    participant_id = os.path.basename(file).split('_')[0]

    # Find the corresponding label information for the participant
    label_info = labels_df.loc[labels_df['participant_id'] == participant_id]

    if not label_info.empty:
        # Extract the class label based on the 'Group' column
        label = class_mapping.get(label_info['Group'].values[0])

        # Read EEG data
        raw = mne.io.read_raw_eeglab(file)
        channel_names = raw.ch_names
        data = raw.get_data(picks=channel_names)
        transposed_data = np.transpose(data)
        data = pd.DataFrame(transposed_data, columns=channel_names)

        # Apply median filtering
        data = data.groupby(data.index // 50).median()

        # Standardize the data
        scaler = StandardScaler()
        data[channel_names] = scaler.fit_transform(data[channel_names])

        # Add the 'Label' column with the assigned class label
        data['Label'] = label

        # Append the processed data to the list
        data_list.append(data)

# Concatenate DataFrames in data_list
concatenated_data = pd.concat(data_list, ignore_index=True)

# Save the concatenated DataFrame to CSV
csv_filepath = os.path.join(output_dir, "datalist_processed.csv")
concatenated_data.to_csv(csv_filepath, index=False)

print(f"Processed concatenated data saved to {csv_filepath}")
```

```python
#2CLASS LSTM ARTICLE BASED MODEL WITH Test accuracy: 0.7222597599029541
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.optimizers import RMSprop,Adam
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras.models import Sequential
original_data = pd.read_csv("/content/drive/MyDrive/dataseta_processed/rpca_results.csv")
X = original_data.iloc[:, 0:-1]
y = original_data['Label'].values

# Filter data to include only class 0 and 1
filter_mask = (y == 0) | (y == 1)
X_filtered = X[filter_mask]
y_filtered = y[filter_mask]

num_classes = 2  # Set the number of classes to 2 for binary classification
X_reshaped = X_filtered.to_numpy().reshape((X_filtered.shape[0], 1, X_filtered.shape[1]))

# Split the filtered data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y_filtered, test_size=0.2, random_state=42)


dense1 = 256
lstm1 = 128
lstm2 = 64
lstm3 = 32
print(X_reshaped.shape)

# Define the function to create the model
def create_model(dense1, lstm1, lstm2, lstm3, dropout=0.3):
    print("\n### Creating model")

    model = Sequential()
    model.add(keras.Input(shape=(X_train.shape[1], X_train.shape[2])))

    if dense1:
        model.add(Dense(dense1, activation='relu'))
        model.add(Dropout(dropout))

    model.add(LSTM(lstm1, return_sequences=True))
    model.add(Dropout(dropout))

    if lstm2:
        model.add(LSTM(lstm2, return_sequences=True))
        model.add(Dropout(dropout))

    if lstm3:
        model.add(LSTM(lstm3))
        model.add(Dropout(dropout))

    model.add(Dense(num_classes, activation='sigmoid'))  # Use 'sigmoid' for binary classification

    model.compile(loss=keras.losses.BinaryCrossentropy(),
                  optimizer=Adam(learning_rate=0.001),
                  metrics=['accuracy'])

    model.summary()
    return model

model = create_model(dense1, lstm1, lstm2, lstm3)

# One-hot encode the labels for training
y_train_onehot = keras.utils.to_categorical(y_train, num_classes)

# Implement early stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
history = model.fit(X_train, y_train_onehot, epochs=100, validation_split=0.2, callbacks=[early_stopping])
def plot_history(history):
    plt.figure(figsize=(12, 4))

    # Plot training & validation accuracy values
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train', 'Validation'], loc='upper left')
    # Plot training & validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.tight_layout()
    plt.show()





# Plot training history
plot_history(history)

# One-hot encode the labels for testing
y_test_onehot = keras.utils.to_categorical(y_test, num_classes)

test_loss, test_acc = model.evaluate(X_test, y_test_onehot)
print(f'Test accuracy: {test_acc}')
```

```python
#APPLY PCA AND RPCA ALGORITM ON DATA
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import sys
sys.path.append('/content/r_pca.py')  # Replace with the actual path to the r_pca library file
from r_pca import R_pca


def pca_reduction(A, tol, comp=0):
    assert len(A.shape) == 2
    dmin = min(A.shape)
    r = R_pca(A)
    L, S = r.fit(tol=tol, max_iter=10, iter_print=1)
    U, lam, V = np.linalg.svd(L, full_matrices=False)
    lam_trunc = lam[lam > 0.015 * lam[0]]
    p = min(comp, dmin) if comp else len(lam_trunc)
    assert p <= dmin
    print('PCA truncation', dmin, '->', p)
    return L, V.T[:, :p]

def reduce_matrix(A):
    B = A.reshape((A.shape[0], -1))
    L, V = pca_reduction(B, 5e-6, comp=50)
    reduced_data = B @ V
    return reduced_data.reshape((A.shape[0], A.shape[1], -1)), V

def adjust_size(x, y):
    if len(x) == len(y):
        return y
    factor = len(x) // len(y)
    ynew = np.empty((len(x), 1))
    for i in range(0, len(y)):
        ynew[i * factor: (i + 1) * factor] = y[i]
    return ynew

dir_path = r'/content/drive/MyDrive/dataseta_processed'
label_file_path = '/content/drive/MyDrive/alzimer_result/participants.csv'

# Load concatenated processed data
concatenated_data = pd.read_csv(os.path.join(dir_path, "datalist_processed.csv"))

# Extract features (EEG data) and labels
features = concatenated_data.drop(columns=['Label'])
labels = concatenated_data['Label'].values

# Apply PCA
original_data = features.values  # Replace this with your actual features
print('\nPerforming (R)(MS)PCA...')
your_data, Vpca = reduce_matrix(original_data)
y_data = adjust_size(your_data, labels)


# Save the results to a new file
pca_results = pd.DataFrame(data=your_data.reshape(your_data.shape[0], -1), columns=concatenated_data.columns[:-1])  # excluding the 'Label' column
pca_results['Label'] = y_data
pca_results.to_csv(os.path.join(dir_path, 'rpca_results.csv'), index=False)
```

```python
#3CLASS LSTM ARTICLE BASED MODEL WITH Test accuracy: 0.6023398637771606
from tensorflow.keras.layers import Dense, LSTM, Dropout, BatchNormalization
from tensorflow.keras.optimizers import RMSprop,Adam
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow.keras as keras
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow import keras
from tensorflow.keras.models import Sequential
from keras import regularizers


original_data = pd.read_csv("/content/drive/MyDrive/dataseta_processed/rpca_results.csv")
X = original_data.iloc[:, 0:-1]
y = original_data['Label'].values

num_classes =3
X_reshaped = X.to_numpy().reshape((X.shape[0], 1, X.shape[1]))

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y, test_size=0.2, random_state=42)



# Define the function to create the model
def create_model(dense1=256, lstm1=128, lstm2=64, lstm3=32, dropout=0.3, learning_rate=0.001):
    model = Sequential()
    model.add(Dense(dense1, activation='relu'))
    model.add(Dropout(dropout))
    model.add(LSTM(lstm1, return_sequences=True))
    model.add(Dropout(dropout))
    model.add(LSTM(lstm2, return_sequences=True))
    model.add(Dropout(dropout))
    model.add(LSTM(lstm3))
    model.add(Dropout(dropout))
    model.add(Dense(num_classes, activation='softmax'))

    model.compile(loss=keras.losses.CategoricalCrossentropy(),
                  optimizer=Adam(learning_rate=learning_rate),
                  metrics=['accuracy'])
    return model

model = create_model()
# Implement early stopping to prevent overfitting
y_train_onehot = keras.utils.to_categorical(y_train, num_classes)
early_stopping = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)

history = model.fit(X_train, y_train_onehot, epochs=100, validation_split=0.2, callbacks=[early_stopping])

def plot_history(history):

    plt.figure(figsize=(12, 4))

    # Plot training & validation accuracy values
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train', 'Validation'], loc='upper left')

    # Plot training & validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.tight_layout()
    plt.show()
# Plot training history
plot_history(history)

# One-hot encode the labels for testing
y_test_onehot = keras.utils.to_categorical(y_test, num_classes)
test_loss, test_acc = model.evaluate(X_test, y_test_onehot)
print(f'Test accuracy: {test_acc}')

```
